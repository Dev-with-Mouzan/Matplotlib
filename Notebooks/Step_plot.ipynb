```python
import matplotlib.pyplot as plt
from matplotlib.typing import MarkerType
import pandas as pd
```

```python
df=pd.read_csv('E:\Data visual\Data\Data.csv')
df.head()
```

```python
top5=df.head()
```

```python

plt.step(top5["Country"],top5["Happiness"],where="post",color="green",marker="*",markersize=15)
plt.grid(True)
plt.title("Step Plot")
plt.xlabel("Country")
plt.ylabel("Happiness")
plt.show()
```
