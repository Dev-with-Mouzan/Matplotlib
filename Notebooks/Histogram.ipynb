```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df=pd.read_csv('E:\Data visual\Data\Data.csv')
df.head()
```

```python
# HistoGram Plot
plt.hist(df["Happiness"],bins=50,align="mid",color="green",histtype="stepfilled",rwidth=0.8,orientation="vertical",log=True)
plt.grid(True)
plt.title("Histogram Plot")
plt.xlabel("Happiness")
plt.ylabel("Frequency")
plt.show()
```
